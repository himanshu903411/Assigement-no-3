```python
# 1. write to code to reverse a string

# Function to reverse a string
def reverse_string(s):
    return s[::-1]

# Example usage
input_string = "hello"
reversed_string = reverse_string(input_string)
print("Original string:", input_string)
print("Reversed string:", reversed_string)

# 2.write to code count the number of vowels in a string

# Function to count the number of vowels in a string
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

# Example usage
input_string = "Hello World"
vowel_count = count_vowels(input_string)
print("Number of vowels:", vowel_count)

# 3. write to code to cheak if give a string is a palindrome.

# Function to check if a string is a palindrome
def is_palindrome(s):
    # Remove spaces and convert to lowercase
    s = s.replace(" ", "").lower()
    # Check if the string is equal to its reverse
    return s == s[::-1]

# Example usage
input_string = "racecar"
if is_palindrome(input_string):
    print(f"'{input_string}' is a palindrome.")
else:
    print(f"'{input_string}' is not a palindrome.")

  # 4. write to code to cheak if two strings are anagrams of each other.

  # Function to check if two strings are anagrams
def are_anagrams(str1, str2):
    # Remove spaces and convert to lowercase
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()

    # Check if sorted versions of both strings are equal
    return sorted(str1) == sorted(str2)

# Example usage
string1 = "listen"
string2 = "silent"

if are_anagrams(string1, string2):
    print(f"'{string1}' and '{string2}' are anagrams.")
else:
    print(f"'{string1}' and '{string2}' are not anagrams.")

# 5. write to code to find all occurrence of a given substring within another string.

# Function to find all occurrences of a substring within a string
def find_all_occurrences(main_string, sub_string):
    occurrences = []
    start = 0

    while start < len(main_string):
        # Find the next occurrence of the substring
        index = main_string.find(sub_string, start)

        # If no occurrence is found, break the loop
        if index == -1:
            break

        # Append the index to the list of occurrences
        occurrences.append(index)

        # Move start index to continue searching
        start = index + len(sub_string)

    return occurrences

# Example usage
main_string = "This is a test, this test is fun"
sub_string = "test"
occurrences = find_all_occurrences(main_string, sub_string)
print(f"Occurrences of '{sub_string}' in '{main_string}': {occurrences}")

# 6. Write to code to perform basic string compression of a given substring within another string.

# Function to perform basic string compression
def compress_string(s):
    compressed = []
    count = 1

    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed.append(s[i - 1] + str(count))
            count = 1

    # Append the last character and its count
    compressed.append(s[-1] + str(count))

    # Join the list into a single compressed string
    compressed_string = ''.join(compressed)

    # Return the original string if compression does not reduce size
    return compressed_string if len(compressed_string) < len(s) else s

# Example usage
input_string = "aabcccccaaa"
compressed_string = compress_string(input_string)
print(f"Compressed string: {compressed_string}")

# 7. Write a code to determine if a string has all unique characters.

# Function to perform basic string compression
def compress_string(s):
    compressed = []
    count = 1

    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed.append(s[i - 1] + str(count))
            count = 1

    # Append the last character and its count
    compressed.append(s[-1] + str(count))

    # Join the list into a single compressed string
    compressed_string = ''.join(compressed)

    # Return the original string if compression does not reduce size
    return compressed_string if len(compressed_string) < len(s) else s

# Example usage
input_string = "aabcccccaaa"
compressed_string = compress_string(input_string)
print(f"Compressed string: {compressed_string}")

# 8. write a code to convert a given string to uppercase or lowercase.

# Function to convert string to uppercase
def to_uppercase(s):
    return s.upper()

# Function to convert string to lowercase
def to_lowercase(s):
    return s.lower()

# Example usage
input_string = "Hello World!"

uppercase_string = to_uppercase(input_string)
lowercase_string = to_lowercase(input_string)

print(f"Original string: {input_string}")
print(f"Uppercase: {uppercase_string}")
print(f"Lowercase: {lowercase_string}")

# 9. Write a code to count the number of the words in a string.

# Function to count the number of words in a string
def count_words(s):
    # Split the string into words using spaces
    words = s.split()
    return len(words)

# Example usage
input_string = "Hello, how are you doing today?"
word_count = count_words(input_string)

print(f"Number of words in the string: {word_count}")

# 10 . Write a code to concatenate two string without using the + operater.

# Function to concatenate two strings
def concatenate_strings(str1, str2):
    return ''.join([str1, str2])

# Example usage
string1 = "Hello"
string2 = "World"
result = concatenate_strings(string1, string2)

print(f"Concatenated string: {result}")

# 11. write to code remove all occurrences of a specific element form a list.

# Function to remove all occurrences of a specific element from a list
def remove_occurrences(lst, element):
    return [x for x in lst if x != element]

# Example usage
original_list = [1, 2, 3, 4, 2, 5, 2, 6]
element_to_remove = 2
modified_list = remove_occurrences(original_list, element_to_remove)

print(f"Original list: {original_list}")
print(f"Modified list (without {element_to_remove}): {modified_list}")

#12.Implement a code to find the second largest number in a given list of integers.

# Function to find the second largest number in a list
def find_second_largest(lst):
    # Initialize variables to store the largest and second largest numbers
    largest = second_largest = float('-inf')

    for number in lst:
        if number > largest:
            second_largest = largest
            largest = number
        elif number > second_largest and number != largest:
            second_largest = number

    # Check if second largest was found
    if second_largest == float('-inf'):
        return None  # or raise an exception for cases with insufficient unique numbers

    return second_largest

# Example usage
numbers = [5, 3, 1, 4, 2, 5, 3, 4]
second_largest = find_second_largest(numbers)

if second_largest is not None:
    print(f"The second largest number in the list is: {second_largest}")
else:
    print("The list does not contain enough unique numbers.")

# 13. Create a code to count the occurrences of each element in a list and return a dictionary with elements as keys and their counts as values.

# Function to count occurrences of each element in a list
def count_occurrences(lst):
    count_dict = {}

    for element in lst:
        if element in count_dict:
            count_dict[element] += 1
        else:
            count_dict[element] = 1

    return count_dict

# Example usage
numbers = [1, 2, 2, 3, 1, 4, 3, 1, 5, 2]
occurrences = count_occurrences(numbers)

print("Occurrences of each element:", occurrences)

# 14. Write a code to reverse a list in-place without using any built-in reverse functions.

# Function to reverse a list in-place
def reverse_list(lst):
    left = 0
    right = len(lst) - 1

    while left < right:
        # Swap the elements at left and right indices
        lst[left], lst[right] = lst[right], lst[left]

        # Move towards the middle
        left += 1
        right -= 1

# Example usage
numbers = [1, 2, 3, 4, 5]
print("Original list:", numbers)

reverse_list(numbers)
print("Reversed list:", numbers)

#.15 Implement a code to find and remove duplicates from a list while preserving the original order of elements.

# Function to remove duplicates while preserving order
def remove_duplicates(lst):
    seen = set()  # Set to keep track of seen elements
    result = []   # List to store the result without duplicates

    for element in lst:
        if element not in seen:
            seen.add(element)  # Add to seen set
            result.append(element)  # Append to result list

    return result

# Example usage
numbers = [1, 2, 3, 1, 2, 4, 5, 3, 6]
unique_numbers = remove_duplicates(numbers)

print("Original list:", numbers)
print("List after removing duplicates:", unique_numbers)

# 16. # Function to check if a list is sorted
def is_sorted(lst):
    if len(lst) < 2:  # A list with less than 2 elements is considered sorted
        return True

    ascending = all(lst[i] <= lst[i + 1] for i in range(len(lst) - 1))
    descending = all(lst[i] >= lst[i + 1] for i in range(len(lst) - 1))

    return ascending or descending

# Example usage
numbers_asc = [1, 2, 3, 4, 5]
numbers_desc = [5, 4, 3, 2, 1]
numbers_unsorted = [1, 3, 2, 5]

print(f"Is the list {numbers_asc} sorted? {is_sorted(numbers_asc)}")         # True
print(f"Is the list {numbers_desc} sorted? {is_sorted(numbers_desc)}")       # True
print(f"Is the list {numbers_unsorted} sorted? {is_sorted(numbers_unsorted)}") # False

#17. Write a code to merge two sorted lists into a single sorted list.

# Function to merge two sorted lists
def merge_sorted_lists(list1, list2):
    merged_list = []
    i, j = 0, 0

    # Traverse both lists and append the smaller element to the merged list
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    # If there are remaining elements in list1, add them
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1

    # If there are remaining elements in list2, add them
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1

    return merged_list

# Example usage
sorted_list1 = [1, 3, 5, 7]
sorted_list2 = [2, 4, 6, 8]
merged = merge_sorted_lists(sorted_list1, sorted_list2)

print("Merged sorted list:", merged)

# 18. Implement a code to find the intersection of two given lists.

# Function to find the intersection of two lists
def find_intersection(list1, list2):
    # Convert one of the lists to a set for efficient lookups
    set2 = set(list2)

    # Use a list comprehension to find common elements
    intersection = [element for element in list1 if element in set2]

    return intersection

# Example usage
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
intersection = find_intersection(list1, list2)

print("Intersection of the two lists:", intersection)

# 19. Create a code to find the union of two lists without duplicates.

# Function to find the union of two lists without duplicates
def find_union(list1, list2):
    # Convert both lists to sets to remove duplicates and perform the union
    union_set = set(list1) | set(list2)

    # Convert the set back to a list
    union_list = list(union_set)

    return union_list

# Example usage
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
union = find_union(list1, list2)

print("Union of the two lists without duplicates:", union)

# 20. Write a code to shuffle a given list randomly without using any built-in shuffle functions

import random

# Function to shuffle a list randomly
def shuffle_list(lst):
    # Create a copy of the list to avoid modifying the original
    shuffled = lst[:]

    # Implement the Fisher-Yates shuffle algorithm
    for i in range(len(shuffled) - 1, 0, -1):
        # Generate a random index from 0 to i
        j = random.randint(0, i)
        # Swap elements at i and j
        shuffled[i], shuffled[j] = shuffled[j], shuffled[i]

    return shuffled

# Example usage
original_list = [1, 2, 3, 4, 5]
shuffled_list = shuffle_list(original_list)

print("Original list:", original_list)
print("Shuffled list:", shuffled_list)

# 21.Write a code that takes two tuples as input and returns a new tuple containing elements that are common to both input tuples.

# Function to find common elements in two tuples
def common_elements(tuple1, tuple2):
    # Convert tuples to sets to find common elements
    set1 = set(tuple1)
    set2 = set(tuple2)

    # Find intersection and convert back to a tuple
    common = set1.intersection(set2)

    return tuple(common)

# Example usage
tuple1 = (1, 2, 3, 4, 5)
tuple2 = (4, 5, 6, 7, 8)
common_tuple = common_elements(tuple1, tuple2)

print("Common elements:", common_tuple)

# 22. Create a code that prompts the user to enter two sets of integers separated by commas. Then, print the intersection of these two sets.

# Function to get user input and find the intersection of two sets
def find_set_intersection():
    # Prompt the user to enter two sets of integers
    set1_input = input("Enter the first set of integers separated by commas: ")
    set2_input = input("Enter the second set of integers separated by commas: ")

    # Convert the input strings to sets of integers
    set1 = {int(x) for x in set1_input.split(",")}
    set2 = {int(x) for x in set2_input.split(",")}

    # Find the intersection of the two sets
    intersection = set1.intersection(set2)

    # Print the result
    print("Intersection of the two sets:", intersection)

# Example usage
find_set_intersection()

# 23.Write a code to concatenate two tuples. The function should take two tuples as input and return a new tuple containing elements from both input tuples.

# Function to concatenate two tuples
def concatenate_tuples(tuple1, tuple2):
    # Concatenate the two tuples using the + operator
    concatenated = tuple1 + tuple2
    return concatenated

# Example usage
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
result = concatenate_tuples(tuple1, tuple2)

print("Concatenated tuple:", result)

# 24.Develop a code that prompts the user to input two sets of strings. Then, print the elements that are present in the first set but not in the second set.

# Function to get user input and find elements in the first set not in the second
def find_difference():
    # Prompt the user to enter two sets of strings
    set1_input = input("Enter the first set of strings separated by commas: ")
    set2_input = input("Enter the second set of strings separated by commas: ")

    # Convert the input strings to sets
    set1 = {x.strip() for x in set1_input.split(",")}
    set2 = {x.strip() for x in set2_input.split(",")}

    # Find the elements present in the first set but not in the second
    difference = set1.difference(set2)

    # Print the result
    print("Elements in the first set but not in the second set:", difference)

# Example usage
find_difference()

# 25. Create a code that takes a tuple and two integers as input. The function should return a new tuple containing elements from the original tuple within the specified range of indices.

# Function to extract elements from a tuple within a specified range of indices
def extract_range_from_tuple(original_tuple, start_index, end_index):
    # Check for valid indices
    if start_index < 0 or end_index > len(original_tuple) or start_index >= end_index:
        raise ValueError("Invalid index range.")

    # Return a new tuple containing elements from start_index to end_index (exclusive)
    return original_tuple[start_index:end_index]

# Example usage
original_tuple = (10, 20, 30, 40, 50, 60)
start = 1
end = 4
result = extract_range_from_tuple(original_tuple, start, end)

print("Extracted tuple:", result)

# 26. Write a code that prompts the user to input two sets of characters. Then, print the union of these two sets.

# Function to get user input and find the union of two sets of characters
def find_union_of_sets():
    # Prompt the user to enter two sets of characters
    set1_input = input("Enter the first set of characters separated by commas: ")
    set2_input = input("Enter the second set of characters separated by commas: ")

    # Convert the input strings to sets of characters
    set1 = {x.strip() for x in set1_input.split(",")}
    set2 = {x.strip() for x in set2_input.split(",")}

    # Find the union of the two sets
    union_set = set1.union(set2)

    # Print the result
    print("Union of the two sets:", union_set)

# Example usage
find_union_of_sets()

# 27.Develop a code that takes a tuple of integers as input. The function should return the maximum and minimum values from the tuple using tuple unpacking.

# Function to find the maximum and minimum values in a tuple
def find_max_min(input_tuple):
    # Check if the tuple is empty
    if not input_tuple:
        raise ValueError("The input tuple is empty.")

    # Using built-in functions to find max and min
    max_value = max(input_tuple)
    min_value = min(input_tuple)

    # Return the max and min values as a tuple
    return max_value, min_value  # Tuple unpacking will happen in the caller

# Example usage
tuple_of_integers = (10, 20, 5, 30, -10)
max_value, min_value = find_max_min(tuple_of_integers)

print("Maximum value:", max_value)
print("Minimum value:", min_value)

# 28. Create a code that defines two sets of integers. Then, print the union, intersection, and difference of these two sets.

# Define two sets of integers
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Calculate union, intersection, and difference
union_set = set1.union(set2)
intersection_set = set1.intersection(set2)
difference_set = set1.difference(set2)

# Print the results
print("Set 1:", set1)
print("Set 2:", set2)
print("Union of Set 1 and Set 2:", union_set)
print("Intersection of Set 1 and Set 2:", intersection_set)
print("Difference (Set 1 - Set 2):", difference_set)

# 29. Write a code that takes a tuple and an element as input. The function should return the count of occurrences of the given element in the tuple.

# Function to count occurrences of an element in a tuple
def count_occurrences(input_tuple, element):
    # Use the count() method to count occurrences
    count = input_tuple.count(element)
    return count

# Example usage
tuple_of_elements = (1, 2, 3, 4, 5, 2, 2, 6)
element_to_count = 2
occurrences = count_occurrences(tuple_of_elements, element_to_count)

print(f"The element {element_to_count} occurs {occurrences} times in the tuple.")

# 30.Develop a code that prompts the user to input two sets of strings. Then, print the symmetric difference of these two sets.

# Function to get user input and find the symmetric difference of two sets of strings
def find_symmetric_difference():
    # Prompt the user to enter two sets of strings
    set1_input = input("Enter the first set of strings separated by commas: ")
    set2_input = input("Enter the second set of strings separated by commas: ")

    # Convert the input strings to sets of strings
    set1 = {x.strip() for x in set1_input.split(",")}
    set2 = {x.strip() for x in set2_input.split(",")}

    # Find the symmetric difference of the two sets
    symmetric_difference = set1.symmetric_difference(set2)

    # Print the result
    print("Symmetric difference of the two sets:", symmetric_difference)

# Example usage
find_symmetric_difference()

# 31. Write a code that takes a list of words as input and returns a dictionary where the keys are unique words and the values are the frequencies of those words in the input list.

# Function to count the frequency of words in a list
def count_word_frequencies(words):
    word_count = {}  # Initialize an empty dictionary

    for word in words:
        # Convert the word to lowercase to ensure case-insensitive counting
        word = word.lower()
        if word in word_count:
            word_count[word] += 1  # Increment count if the word is already in the dictionary
        else:
            word_count[word] = 1   # Initialize count if it's the first occurrence

    return word_count

# Example usage
input_words = input("Enter a list of words separated by spaces: ").split()
word_frequencies = count_word_frequencies(input_words)

print("Word Frequencies:", word_frequencies)

# 32.Write a code that takes two dictionaries as input and merges them into a single dictionary. If there are common keys, the values should be added together.

# Function to merge two dictionaries
def merge_dictionaries(dict1, dict2):
    merged_dict = dict1.copy()  # Create a copy of the first dictionary

    for key, value in dict2.items():
        if key in merged_dict:
            merged_dict[key] += value  # Add values for common keys
        else:
            merged_dict[key] = value  # Add new key-value pair

    return merged_dict

# Example usage
dict1 = {'a': 10, 'b': 20, 'c': 30}
dict2 = {'b': 15, 'c': 25, 'd': 5}

merged_result = merge_dictionaries(dict1, dict2)

print("Merged Dictionary:", merged_result)

# 33.Write a code to access a value in a nested dictionary. The function should take the dictionary and a list of keys as input, and return the corresponding value. If any of the keys do not exist in the dictionary, the function should return None.

# Function to access a value in a nested dictionary using a list of keys
def access_nested_value(nested_dict, keys):
    current_value = nested_dict  # Start with the whole dictionary

    for key in keys:
        # Check if the key exists in the current level of the dictionary
        if isinstance(current_value, dict) and key in current_value:
            current_value = current_value[key]  # Move deeper into the dictionary
        else:
            return None  # Key does not exist

    return current_value  # Return the final value

# Example usage
nested_dictionary = {
    'a': {
        'b': {
            'c': 42
        }
    },
    'd': 100
}

keys_to_access = ['a', 'b', 'c']
value = access_nested_value(nested_dictionary, keys_to_access)

if value is not None:
    print(f"The value for the keys {keys_to_access} is: {value}")
else:
    print(f"The keys {keys_to_access} do not exist in the dictionary.")

# 34. Write a code that takes a dictionary as input and returns a sorted version of it based on the values. You can choose whether to sort in ascending or descending order.

# Function to sort a dictionary by its values
def sort_dictionary_by_values(input_dict, descending=False):
    # Sort the dictionary by values and return as a list of tuples
    sorted_dict = sorted(input_dict.items(), key=lambda item: item[1], reverse=descending)

    # Convert the sorted list of tuples back into a dictionary
    return dict(sorted_dict)

# Example usage
input_dictionary = {'apple': 5, 'banana': 2, 'cherry': 10, 'date': 1}

# Sort in ascending order
sorted_ascending = sort_dictionary_by_values(input_dictionary)
print("Sorted Dictionary (Ascending):", sorted_ascending)

# Sort in descending order
sorted_descending = sort_dictionary_by_values(input_dictionary, descending=True)
print("Sorted Dictionary (Descending):", sorted_descending)

# 35.Write a code that inverts a dictionary, swapping keys and values. Ensure that the inverted dictionary correctly handles cases where multiple keys have the same value by storing the keys as a list in the inverted dictionary.

# Function to invert a dictionary
def invert_dictionary(input_dict):
    inverted_dict = {}  # Initialize an empty dictionary for the inverted result

    for key, value in input_dict.items():
        # Check if the value already exists in the inverted dictionary
        if value in inverted_dict:
            # Append the key to the existing list of keys
            inverted_dict[value].append(key)
        else:
            # Create a new entry with the value as the key and the key as the first element of the list
            inverted_dict[value] = [key]

    return inverted_dict

# Example usage
input_dictionary = {
    'a': 1,
    'b': 2,
    'c': 1,
    'd': 3,
    'e': 2
}

inverted_result = invert_dictionary(input_dictionary)

print("Inverted Dictionary:", inverted_result)












```
